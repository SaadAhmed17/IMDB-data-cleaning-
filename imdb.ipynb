### Importing libraries (pandas and numpy) and load dataset

```python
import pandas as pd
import numpy as np
df = pd.read_excel('IMDB.xlsx')
df
```

### Remove duplicate rows and strip whitespace from column names

```python
df.drop_duplicates()
df.columns = df.columns.str.strip()
```

### Replace spaces in column names with underscores to make them code-friendly

```python
df.columns = df.columns.str.replace(' ', '_')
```

### Capitalize column names and fix specific typos/special characters in 'Original_titlê' and 'Genrë¨' column names

```python
df.columns = df.columns.str.capitalize()
df.columns = [col.replace('ê','e').replace('ë', 'e').replace('¨','') if col in ['Original_titlê', 'Genrë¨'] else col for col in df.columns]
df.head(16)
```

### Drop rows where all column values are missing (completely empty rows)

```python
df.dropna(how='all', inplace=True)
df
```

### Remove unwanted punctuation and special characters (: , . · ¹) from the 'Original_title' column using regex

```python
df["Original_title"] = df["Original_title"].str.replace(r"[:,.·¹]",'', regex = True)
df.head()
```

### Replace specific incorrect or misspelled movie titles in the 'Original_title' column using a custom dictionary


```python
replace_dict = {
    'Se7en': 'Seven',
    "Le fabuleux destin d'AmÃ©lie Poulain": "Le fabuleux destin damÃ©lie Poulain",
    'Per qualche dollaro in piA': 'Per qualche dollaro in pia'

} 
df['Original_title'] = df['Original_title'].replace(replace_dict)
```

### Fix character encoding issues in 'Original_title' by replacing 'Ã' with 'A' and '©' with 'c'

```python
df['Original_title'] = (df['Original_title']
.str.replace('Ã', 'A', regex = False)
.str.replace('©', 'c', regex = False))
```

### Parse and clean various date formats in 'Release_year', convert unparseable values to NaT, and drop rows with invalid dates


```python
from dateutil import parser
def clean_date(value):
    try:
        # Try to parse the date
        return parser.parse(value, fuzzy=True).date()
    except:
        # Return NaT (Not a Time) if invalid
        return pd.NaT

df['Release_year'] = df['Release_year'].astype(str).apply(clean_date)

# See rows where date could not be parsed
invalid_dates = df[df['Release_year'].isna()]

# Drop invalid rows if needed
df = df.dropna(subset=['Release_year'])
df.head(50)
```

### Convert all values in the 'Genre' column to lowercase and remove leading/trailing whitespace


```python
df['Genre'] = df['Genre'].str.lower().str.strip()
print(df['Genre'].head(10))
```

### Split the 'Genre' column into lists of genres using comma as the separator, allowing for optional spaces


```python
df['Genre'] = df['Genre'].str.split(r',\s*')
df
```

### Explode the lists in the 'Genre' column so each genre appears in its own row, then count the frequency of each genre


```python
df.explode('Genre')
print(df.explode('Genre')['Genre'].value_counts())
```

### Strip whitespace from 'Duration' values and replace specific invalid entries (e.g., 'Inf', 'Not Applicable', empty strings) with cleaned values or '0'


```python
df['Duration'].str.strip()
replac_dict = {
    'Inf': '',
    '178c': '178',
    'Not Applicable': '',
    '-': '',
    '': '0'
}
df['Duration'] = df['Duration'].replace(replac_dict)
```

```python
df['Duration'] = df['Duration'].astype(str).str.strip()
```

```python
df['Duration'] = df['Duration'].replace(['', 'Nan'], '0')
df.head(20)
```

### Convert 'Duration' to numeric, calculate the mean of non-zero durations, and replace 0 values with the calculated mean

```python
df['Duration'] = pd.to_numeric(df['Duration'])
duration_mean = round(df[df['Duration'] != 0]['Duration'].mean())
duration_mean
df['Duration'] = df['Duration'].replace(0, duration_mean)
df.head(20)
```

### Remove trailing periods from 'Country' values and correct misspelled or inconsistent country names using a replacement dictionary


```python
df['Country'] = df['Country'].str.strip('.')
country_dict = {
    'US': 'USA',
    'New Zesland': 'New Zealand',
    'New Zeland': 'New Zealand',
    'Italy1': 'Italy'
}
df['Country'] = df['Country'].replace(country_dict)
```

### Clean 'Content_rating' by stripping whitespace, filling NaN values with 'Unknown', and standardizing inconsistent ratings using a replacement dictionary


```python
df['Content_rating'].dropna()
df['Content_rating'].str.strip()
df['Content_rating'].isna().sum()

cr_dict = {
    'Unrated': 'Not Rated',
    '#N/A': 'Unknown',
    '': 'Unknown',            
    'Approved': 'Not Rated',
   }
df['Content_rating'] = df['Content_rating'].fillna('Unknown')
df['Content_rating'] = df['Content_rating'].replace(cr_dict)
print(df['Content_rating'].value_counts())
df.head()
```

## Director Column

```python
df['Director'] = df['Director'].str.lower().str.strip()
# df['Director'].head(30)
```

### Convert 'Director' values to string, fix specific encoding issues using a replacement dictionary, and capitalize the first letter of each entry


```python
df['Director'] = df['Director'].astype(str)
director_dict = {
    'fernando meirelles, kã¡tia lund': 'fernando meirelles, katia lund',
    'olivier nakache, ã‰ric toledano': 'olivier nakache, aric toledano'
}
df['Director'] = df['Director'].replace(director_dict)
df['Director'] = df['Director'].str.capitalize()
df
```

### Clean 'Income' by removing unwanted characters and converting to numeric, then calculate and round the mean income


```python
df['Income'] = pd.to_numeric(df['Income'].astype(str).str.replace('o', '0').str.replace('$ ', '').str.replace(',', ''))
round(df['Income'].mean())
```

### Remove dot separators from 'Votes' and convert the column to numeric format

```python
df['Votes'] = pd.to_numeric(df['Votes'].astype(str).str.replace('.', ''))
```

### Clean 'Score' by fixing formatting issues (e.g., commas, extra symbols, malformed decimals), then convert to numeric for analysis


```python
df['Score'] = df['Score'].astype(str).str.strip()

df['Score'] = pd.to_numeric(df['Score'].str.replace(',', '.').str.replace('f', '').str.replace('e-', '').str.replace(':', '').str.replace('0', '').str.replace('..', '.0').str.replace('88', '8.8').str.rstrip('.'))

df['Score'].head(23)
```
